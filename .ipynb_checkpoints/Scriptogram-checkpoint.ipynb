```python
import sys
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime

def getdf_fromsub(filename):
    i=0
    data =[]
    # Process with opening the input file.
    with open(filename) as f_in:
        #!remove blank lines!
        lines = filter(None, (line.rstrip() for line in f_in))
        skipfirst=True
        time=""
        text=""
        #remove unnecessary lines and write it to a new file
        for line in lines:
            if skipfirst:
                skipfirst=False
                continue
            # Ignore the numeric ID of each text
            if line.isdigit() and len(line) < 6:
                continue
            # store the time indicator such as '00:00:12,340' 
            if line.startswith('0'):
                if text!="":
                    data.append([time,text])
                time=line[:12]
                text=""
                i+=1
                continue
            # Ignore some trivial attributes such as '<font>~</font>'
            if line.startswith('<font'):
                continue

            #remove <> tags
            line = line.replace("<i>","").replace("</i>","")
            text+=line
    if text!="":
            data.append([time,text])
    
    
    # Calcul du nombre de mots
    df=pd.DataFrame(data=data,columns=['time','text'])
    df['time'] = pd.to_datetime(df['time'])
    df['wordcount']=df['text'].apply(lambda x:len(x.split(" ")))
    df['second']=df['time'].apply(lambda x:int(x.second/10)+x.minute*6+x.hour*360)
    dfg = df.groupby("second")["wordcount"].sum().reset_index()
    
    # Affichage du scriptogramme
    plt.figure(num=None, figsize=(22, 3), dpi=80, facecolor='w', edgecolor='k')
    plt.bar(dfg['second'],dfg['wordcount'],width=2)
    wc=dfg['wordcount'].sum()
    rt=dfg['second'].max()*10
    plt.title("Distribution of {} words spoken for movie {} by 10 sec time period, avg={} words/sec".
              format(wc,filename,np.around(wc/rt,decimals=2)))
    axes = plt.gca()
    axes.set_xlim(0,1250)
    axes.set_ylim(0,60)
    return plt
```
